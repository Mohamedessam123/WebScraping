```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin  # ✅ import urljoin

root = "https://subslikescript.com"
url = f"{root}/movies"


# Step 1: Get the movie links from the main page
result = requests.get(url)
content = result.text
soup = BeautifulSoup(content, "lxml")

all_films = soup.find("article", class_="main-article")
links = []

for link in all_films.find_all("a", href=True):
    links.append(link["href"])

# Step 2: Loop through each movie link
for link in links:
    try:
        # ✅ This fixes the "//" issue
        movie_url = urljoin(root, link)

        result = requests.get(movie_url)
        if result.status_code != 200:
            print(f"❌ Skipping (status {result.status_code}): {movie_url}")
            continue

        soup = BeautifulSoup(result.text, "lxml")
        article = soup.find("article", class_="main-article")

        if not article:
            print(f"❌ No article found at: {movie_url}")
            continue

        title = article.find("h1").get_text()
        safe_title = "".join(c for c in title if c.isalnum() or c in (" ", "_")).strip()

        transcript_div = article.find("div", class_="full-script")
        if transcript_div:
            transcript = transcript_div.get_text(strip=True, separator=" ")
            with open(f"{safe_title}.txt", "w", encoding="utf-8") as file:
                file.write(transcript)
            print(f"✅ Saved: {safe_title}")
        else:
            print(f"❌ Transcript not found for: {title}")

    except Exception as e:
        print(f"⚠️ Error processing {link}: {e}")

```
